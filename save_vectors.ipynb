```python
from pymongo import MongoClient
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.document_loaders import PyPDFLoader
```

```python
emb_model = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
loader = PyPDFLoader("assets/NetSol_Financial Statement_2024_Part 1.pdf")
client = MongoClient('mongodb://localhost:27017/')
# client = MongoClient('mongodb+srv://rreigns273:RuFXC4nd0SZux1F5@cluster0.lsf0m.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0')
```

```python
# Create a database object
db = client['netsol_db']

collection = db['financial_statements']
```

```python
text = ''
async for page in loader.alazy_load():
    text += page.page_content
print(text)
```

```python
def chunk_generator(text, chunk_size=1000, overlap=100):

    start = 0
    while start < len(text):
        chunk = text[start:start+chunk_size]
        start += chunk_size - overlap
        yield chunk
```

```python
for counter, chunk in enumerate(chunk_generator(text, chunk_size=2000, overlap=100)):
    print(f'Processing chunk {counter}')
    embeddings = emb_model.embed_query(chunk)
    data = {
        'id': str(counter),
        'values': embeddings,
        'metadata': {'text': chunk}
    }
    collection.insert_one(data)
```
